```python
##Running loop for conversion of .abi files to fasta while creating a record for each file.

from Bio import SeqIO
import os

# Setting paths for the input and output file
input_dir = 'C:/Users/msaim/Downloads/sample'
output_file ='C:/Users/msaim/Downloads/MSBT-4'

# Loop for iterating through each .abi file in the sample folder
for filename in os.listdir(input_dir):
    if filename.endswith('.abi'):
#Reading in the .abi files and creating a SeqRecord object.
        record = SeqIO.read(os.path.join(input_dir, filename), 'abi')
        
#Writing the created record to the output fasta file.
        with open(output_file, 'a') as f:
            SeqIO.write(record, f, 'fasta')




```

```python
##Viewing the created fasta files.
from Bio import SeqIO

# The path of the output file
fasta_file ='C:/Users/msaim/Downloads/MSBT-4'

# Reading in the FASTA file using SeqIO
records = SeqIO.parse(fasta_file, 'fasta')

# Iterating over each record inorder to print the ID and sequence
for record in records:
    print(f'ID: {record.id}\nSequence: {record.seq}\n')
```
